```python
# 1. Library Book Management System

def library_management_system():
    books = []

    print("--- Welcome to the Library Management System ---")

    # ADDING BOOKS
    try:
        num_to_add = int(input("How many books would you like to add? "))
    except ValueError:
        print("Invalid input. Please enter a number next time.")
        return

    for i in range(num_to_add):
        print(f"\nEntering details for Book #{i + 1}:")
        
        title = input("Enter Title: ").strip()
        author = input("Enter Author: ").strip()
        status_input = input("Is it available? (yes/no): ").lower().strip()
        
        is_available = True if status_input == "yes" else False
        
        books.append({
            "title": title,
            "author": author,
            "available": is_available
        })

    # MAIN MENU LOOP
    while True:
        print("\n" + "="*30)
        print("1. Display All Books")
        print("2. Borrow a Book")
        print("3. Return a Book")
        print("4. Exit")
        choice = input("Select an option (1-4): ")

        # DISPLAY LIST OF BOOKS
        if choice == "1":
            print("\n--- Current Library Inventory ---")
            if not books:
                print("The library is currently empty.")
            for b in books:
                status = "Available" if b["available"] else "Borrowed"
                print(f"Title: {b['title']} | Author: {b['author']} | Status: {status}")

        # BORROW A BOOK
        elif choice == "2":
            target_title = input("Enter the title of the book to borrow: ").strip()
            found = False
            
            for b in books:
                if b["title"].lower() == target_title.lower():
                    found = True
                    if b["available"]:
                        b["available"] = False
                        print(f"SUCCESS: You have successfully borrowed '{b['title']}'.")
                    else:
                        print(f"ERROR: '{b['title']}' is already borrowed.")
                    break
            
            if not found:
                print("ERROR: Book title not found in the system.")

        # RETURN A BOOK
        elif choice == "3":
            target_title = input("Enter the title of the book to return: ").strip()
            found = False
            
            for b in books:
                if b["title"].lower() == target_title.lower():
                    found = True
                    if not b["available"]:
                        b["available"] = True
                        print(f"SUCCESS: You have successfully returned '{b['title']}'.")
                    else:
                        print(f"NOTE: This book was already marked as available.")
                    break
            
            if not found:
                print("ERROR: This book does not belong to our library.")

        elif choice == "4":
            print("Goodbye!")
            break
        else:
            print("Invalid selection. Please choose 1-4.")

# RUN
if __name__ == "__main__":
    library_management_system()
```

```python
# 2. Employee Payroll System

def payroll_system():
    print("--- Employee Payroll System ---")

    # INPUT
    while True:
        try:
            num_employees = int(input("Enter the number of employees: "))
            if num_employees < 1:
                print("Error: You must enter at least one employee.")
            else:
                break
        except ValueError:
            print("Invalid input. Please enter a whole number.")

    employee_list = []

    for i in range(num_employees):
        print(f"\nSetting up details for Employee #{i+1}:")
        name = input("Enter Name: ").strip()

        while True:
            try:
                hourly_wage = float(input(f"Enter hourly wage for {name}: "))
                if hourly_wage < 0:
                    print("Error: Hourly wage cannot be negative.")
                else:
                    break
            except ValueError:
                print("Invalid input. Please enter a number for the wage.")

        while True:
            try:
                hours_worked = float(input(f"Enter total hours worked by {name}: "))
                if hours_worked < 0:
                    print("Error: Hours worked cannot be negative.")
                else:
                    break
            except ValueError:
                print("Invalid input. Please enter a number for hours.")

        # CALCULATIONS
        overtime_threshold = 160
        overtime_multiplier = 1.5
        
        if hours_worked > overtime_threshold:
            base_salary = overtime_threshold * hourly_wage
            overtime_hours = hours_worked - overtime_threshold
            overtime_pay = overtime_hours * (hourly_wage * overtime_multiplier)
        else:
            base_salary = hours_worked * hourly_wage
            overtime_pay = 0.0

        employee_list.append({
            "name": name,
            "hours": hours_worked,
            "base": base_salary,
            "ot": overtime_pay,
            "total": base_salary + overtime_pay
        })

    # OUTPUT
    print("\n" + "="*60)
    print(f"{'NAME':<15} {'HOURS':<10} {'BASE SALARY':<15} {'OVERTIME PAY':<15}")
    print("-" * 60)
    
    for emp in employee_list:
        print(f"{emp['name']:<15} {emp['hours']:<10.1f} ${emp['base']:<14.2f} ${emp['ot']:<14.2f}")
    
    print("="*60)

if __name__ == "__main__":
    payroll_system()
```

```python
# 3. Online Shopping Cart System

def shopping_cart_system():
    inventory = [
        {"name": "Laptop", "price": 45000.0, "stock": 5},
        {"name": "Mouse", "price": 500.0, "stock": 20},
        {"name": "Keyboard", "price": 1200.0, "stock": 10},
        {"name": "Monitor", "price": 7500.0, "stock": 8}
    ]
    
    cart = []

    print("--- Welcome to the Online Store ---")

    while True:
        print("\n" + "="*30)
        print("1. View Products & Add to Cart")
        print("2. Remove Item from Cart")
        print("3. View Cart & Total")
        print("4. Checkout & Exit")
        choice = input("Select an option (1-4): ")

        # ADD PRODUCTS
        if choice == "1":
            print("\nAvailable Products:")
            for i, item in enumerate(inventory):
                print(f"{i + 1}. {item['name']} - ₱{item['price']} (Stock: {item['stock']})")
            
            try:
                selection = int(input("\nEnter the product number to add: ")) - 1
                if 0 <= selection < len(inventory):
                    product = inventory[selection]
                    
                    # Prevent adding more than available stock
                    if product['stock'] > 0:
                        qty = int(input(f"How many {product['name']}s? "))
                        if 0 < qty <= product['stock']:
                            # Update stock and add to cart
                            product['stock'] -= qty
                            cart.append({"name": product["name"], "price": product["price"], "qty": qty})
                            print(f"SUCCESS: Added {qty} {product['name']}(s) to cart.")
                        else:
                            print(f"ERROR: Invalid quantity. Available stock: {product['stock']}")
                    else:
                        print("ERROR: Out of stock!")
                else:
                    print("ERROR: Invalid selection.")
            except ValueError:
                print("ERROR: Please enter a valid number.")

        # REMOVE PRODUCTS
        elif choice == "2":
            if not cart:
                print("Your cart is empty.")
                continue
            
            print("\nYour Current Cart:")
            for i, item in enumerate(cart):
                print(f"{i + 1}. {item['name']} (Qty: {item['qty']})")
            
            try:
                remove_idx = int(input("\nEnter the item number to remove: ")) - 1
                if 0 <= remove_idx < len(cart):
                    removed_item = cart.pop(remove_idx)
                    
                    # Return stock to inventory
                    for product in inventory:
                        if product['name'] == removed_item['name']:
                            product['stock'] += removed_item['qty']
                            break
                    
                    print(f"SUCCESS: Removed {removed_item['name']} from cart.")
                else:
                    print("ERROR: Invalid selection.")
            except ValueError:
                print("ERROR: Please enter a number.")

        # CALCULATIONS AND OUTPUT
        elif choice == "3":
            print("\n--- Your Shopping Cart ---")
            if not cart:
                print("Cart is empty.")
            else:
                total_price = 0
                for item in cart:
                    item_total = item['price'] * item['qty']
                    total_price += item_total
                    print(f"- {item['name']} (x{item['qty']}): ₱{item_total:,.2f}")
                print("-" * 25)
                print(f"TOTAL PRICE: ₱{total_price:,.2f}")

        # EXIT
        elif choice == "4":
            print("Thank you for shopping!")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    shopping_cart_system()
```
